```python
import plotly.plotly as py
import plotly.graph_objs as go 
import plotly
import numpy as np
plotly.tools.set_credentials_file(username='XiangYang', api_key='ODNI1EP72cNbMNrDJImP')
```

```python
fig = py.get_figure("dfreder1", 69)

data = fig.get_data()

di = {}

for i in data[0]["x"]:
    if i in di.keys():
        di[i] += 1
    else:
        di[i] = 1
        
new_x = [di.keys()[i] for i in range(len(di.keys())) if di.keys()[i] > 1900]
old_y = [di.values()[i] for i in range(len(di.keys())) if di.keys()[i] > 1900]
new_y = np.cumsum(np.array(old_y))
trace = go.Scatter(x= new_x, y = new_y, 
                   marker = {"color": "orange"}, 
                   name = "Bridges", 
                   fill="tozeroy", fillcolor = "rgba(255, 165, 0, 0.5)")
    
layout = go.Layout(title = "Total Bridges Built in CA Since 1900", 
                   width = 600,
                   height = 400,
                   xaxis = dict(title="Year", dtick = 10), 
                   yaxis = dict(title = "Total Bridges", dtick = 5000),
                   titlefont = {"family": "Times New Roman"})

data = [trace]

figure = go.Figure(data = data, layout = layout)
    
py.iplot(figure, filename= "Python Project")
```

```python
py.image.save_as(figure, filename='Total-Bridges-Built-In-CA-Since-1900.png', scale = 3)
```
