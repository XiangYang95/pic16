```python
import plotly.plotly as py
import plotly.graph_objs as go 
fig = py.get_figure("dfreder1", 69)

layout = go.Layout(title = "Total Bridges Built in CA Since 1900", xaxis = go.XAxis(title = "Years"),
                  yaxis = go.YAxis(title = "Total Bridges"))
trace = go.Scatter(x= fig.data[0].x, y= fig.data["y"],width = 600, height = 400, fill = {"color":"orange"}, 
        marker = {"color": "orange"}, name = "Years")
    
data = go.Data(trace)

figure = go.Figure(data = data, layout = layout)
    
py.iplot(figure, filename= "Python Project")
```
