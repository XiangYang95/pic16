```python
from PyQt5.QtWidgets import QApplication, QWidget, QColorDialog, QRadioButton,QVBoxLayout
from PyQt5.QtGui import QPainter, QColor
class Example(QWidget):

    def __init__(self):
        super(Example, self).__init__()        
        self.initUI()
        
    def initUI(self): 
        b1 = QRadioButton("Button 1")
        b2 = QRadioButton("Button 2")
        qv = QVBoxLayout()
        #initializing the square's dimension
        self.x = 20; self.y = 20
        self.d = 50
        #self.b1.toggled.connect(lambda:self.btnstate(self.b1))
        #initializing the color
        self.qc = QColor(255,0,0)
        
        #adding the buttons onto the widget
        qv.addWidget(b1)
        qv.addStretch()
        qv.addWidget(b2)
        qv.addStretch()
        self.setLayout(qv)
        
        #initializing the widget's dimension
        self.setGeometry(300, 300, 600, 400)
        self.setWindowTitle('Square')
        self.show()
    
    def paintEvent(self, e):
        qp = QPainter()
        qp.begin(self)
        self.drawRectangles(qp)
        qp.end()
        
    def drawRectangles(self, qp):
        qp.setPen(QColor(0, 0, 0, 255))
        
        #draw the widget
        qp.setBrush(QColor(255, 255, 255))
        qp.drawRect(0, 0, self.width(), self.height())
        
        #draw the square
        qp.setBrush(self.qc)
        qp.drawRect(self.x, self.y, self.d, self.d)
def main():
    
    app = QApplication([])
    ex = Example()
    app.exec_()


if __name__ == '__main__':
    main()
```

```python
import scipy.integrate as spint
from sympy import *
from sympy.utilities import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = Symbol("x")


int_f = lambdify(x, integrate(x*sin(x**2)))
diff_f = lambdify(x, diff(x*sin(x**2)))
print int_f(np.array([2,3])) 

x = np.linspace(0, 2*np.pi, 1000)
delta = x[1] - x[0]

print delta
df = diff_f(x)
F = int_f(x)
plt.plot(df, F)
df[0] -= int_f(-delta)/delta**2
df[-1] -= int_f(2*np.pi * delta)/delta**2

def diff2_mat(N):
    j =  np.diag(2*np.ones((N)))-np.diag(np.ones((N-1)), k=-1)-np.diag(np.ones((N-1)), k=1)
    return j

img= diff2_mat(10)

#plt.imshow(img)

D2 = -diff2_mat(1000)/delta**2
print D2[:3,0]
```

```python
import scipy.linalg as lin

F2 = lin.solve(D2, df)
print F2[:2],F[:2]
np.allclose(F2, F, atol = 1)
```

```python
j2 = np.zeros((1000))
j2[0], j2[1] = D2[0][0], D2[1][0]
F2 = lin.solve(lin.toeplitz(j2, j2), df)
np.allclose(F2, F, atol = 1)
```
