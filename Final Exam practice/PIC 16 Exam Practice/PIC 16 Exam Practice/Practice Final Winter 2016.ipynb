```python
import scipy.integrate as spint
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(y,t,m):
    dy = np.zeros(len(y))
    dy[0] = y[1]
    dy[1] = m*(1-y[0]**2)*y[1] - y[0]
    return dy

m = 0.1

f1 = lambda y,t:f(y,t,m)

init_cond = (0.1,0.1)

t = np.linspace(0, 100, 50)

sol = spint.odeint(f1, init_cond, t)

plt.plot(sol[:,0], sol[:,1])
```

```python
from scipy.optimize import minimize
import numpy as np
def f(x,a,b):
    return (a-x[0])**2+b*(x[1]-x[0]**2)**2
a,b = 1,100
f1 = lambda x:f(x,a,b)

x0 = [0,0]
bnds = ([0,0.75],[0,0.75])


def const(x,a,b,r):
    return (x[0]-a)**2+(x[0]-b)**2-r**2

a,b,r = 0.25,0.25,0.25
const1 = lambda x: const(x,a,b,r)

constraint = {"type":"eq", "fun":const1}

res0 = minimize(f1,x0=x0, bounds = bnds)
res1 = minimize(f1,x0=x0, constraints = constraint)

res1
```
