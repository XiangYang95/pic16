```python
img = cv2.imread("c.png",0)

cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("a.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # convert to grayscale

# threshold (binarize) image at intensity 200 out of 255. Invert so that letters are white; paper is black.
ret, thresh = cv2.threshold(gray,200,255,cv2.THRESH_BINARY_INV)

# Alternate way of thresholding the image
#i = np.array(gray < 200)
#thresh = gray*0
#thresh[i] = 255

cv2.imshow("image",thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
im2, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE) # find contours

# The cv2.RETR_EXTERNAL mode is useful here because it only gets the outer contours
# im2, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
# But I didn't find out about it until after writing this, so I don't expect you would!

print len(contours)
cv2.drawContours(img, contours, -1, (0,255,0), 2) # draw them
cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

59 contours are found, but I know there are only 23 letters. Unfortunately, some of the contours are not outer letter boundaries but stray marks or inner letter boundaries. One way to filter these out is based on area. Only the large ones will be outer letter boundaries. However, we need to figure out how large is large enough.

<i>Note:</i> We could have used the <tt>cv2.RETR_EXTERNAL</tt> mode, which only returns <i>outer</i> boundaries of the letters, but we'd still have to eliminate stray marks.

```python
rects = []
areas = []
for contour in contours[::-1]:            # go through in reverse order (not conceptually important)
    boundRect = cv2.boundingRect(contour) # find bounding rectangle (top left x, top left y, width, height) around contour
    area = boundRect[2]*boundRect[3]      # calculate area
    areas.append(area)
    rects.append(boundRect)
```

```python
import matplotlib.pyplot as plt
plt.hist(areas)
# rectangles with areas < 600 are stray marks and the inside of letters
# rectangls with areas > 600 are bounding rectangles of individual letters
```

It turns out that these boundaries work for letters a, b, and c. Area is probably enough to discern between stray marks and letters, but it is possible that the inner boundaries of some letters (capital O, for instance) could be larger than the outer boundaries of other letters (lowercase i, for instance). Then it would be difficult to determine a size threshold that distinguishes between inner and outer boundaries for all letters. One solution would be to throw out very small (<100) contours, assuming they are stray marks, and also eliminate contours that lie entirely within other contours (inner letter boundaries).

```python
# Extract individual letters
letters = []

for area,rect in zip(areas,rects):
    if area > 600:
        x0,y0,w,h = rect
        letter = gray[y0:y0+h,x0:x0+w]
        letters.append(letter)

for letter in letters:
    cv2.imshow("image",letter)
    key = cv2.waitKey(0)
    if key == ord('q'):
        break
            
cv2.destroyAllWindows()
```

```python
# combine all the stuff above into a function that loads image, 
# separates into letters, resizes them, and flattens them to produce a
# single data array with dimensions (n_letters, n_pixels = 25)
def getData(filename):
    img = cv2.imread(filename)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray,200,255,cv2.THRESH_BINARY_INV)
    im2, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    letters = []
    for contour in contours[::-1]:
        rect = cv2.boundingRect(contour)
        area = rect[2]*rect[3]
        if area < 600:
            continue
        x0,y0,w,h = rect
        letter = gray[y0:y0+h,x0:x0+w]
        small = cv2.resize(letter,(5,5))
        letters.append(small.flatten())
    return np.vstack(letters)

def getAllData(filenames):
    data = []
    target = []
    for i,filename in enumerate(filenames):
        d = getData(filename)
        t =[i]*len(d)
        data.append(d)
        target.append(t)
    return np.vstack(data), np.array(target).flatten()

filenames = ["a.png", "b.png", "c.png"]
data, target = getAllData(filenames)
```

```python
def partition(data,target,p):
    n_samples, n_features = data.shape
    i = np.random.permutation(np.arange(n_samples))
    n_train = int(p*n_samples)
    i_train = i[:n_train]
    i_test = i[n_train:]
    return data[i_train], target[i_train], data[i_test], target[i_test]
    
X_train, y_train, X_test, y_test = partition(data,target,0.2)
```

```python
from sklearn.svm import LinearSVC
s = LinearSVC()
s.fit(X_train,y_train)
predicted  = s.predict(X_test)
accuracy = float(sum(y_test == predicted))/len(y_test)
```

```python
print "Predicted:", predicted
print "Truth:", y_test
print "Accuracy: ", accuracy*100, "%"
```

```python
from sklearn.svm import SVC
s2 = SVC(gamma = 0.001) # smaller (0.00001) seems to be better
s2.fit(X_train,y_train)
predicted  = s2.predict(X_test)
accuracy = float(sum(y_test == predicted))/len(y_test)
print "Predicted:", predicted
print "Truth:", y_test
print "Accuracy: ", accuracy*100, "%"
```
