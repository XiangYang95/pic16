```python
A = wav[1][:].max()
data1 = []
for t in np.linspace(0,2,wav[0]*2):
    data1.append(A*np.sin(440*2*math.pi*t))
    
data = np.array(data1)
```

```python
data2 = data.astype("int16")
wf.write("AMajor1.wav", int(44100), data2)
```

```python
A = wav[1][:].max()
anew  = np.array([A*np.sin(440*2*np.pi*t) for t in np.linspace(0,2,wav[0]*2)], dtype = 'int16')
wf.write("A440Mine.wav",44100,anew)
```
