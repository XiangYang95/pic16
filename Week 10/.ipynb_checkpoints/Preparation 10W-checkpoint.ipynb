```python
import matplotlib.pyplot as plt
import numpy as np
clf = svm.SVC(gamma = 0.001, C= 100)
digits = datasets.load_digits()
def accuracies(i):
    train_data = digits.data[:i]
    train_target = digits.target[:i]
    test_data = digits.data[i:]
    test_target = digits.target[i:]
    
    clf.fit(train_data, train_target)
    expected_test_target = clf.predict(test_data)
    n = len(test_target)
    return sum(clf.predict(test_data)==test_target)*1.0/n

l = len(digits.target)
m = range(100,l,10)
p = [accuracies(i) for i in m]
plt.plot(m,p)
```

```python
from sklearn import svm, datasets
digits = datasets.load_digits()
clf = svm.SVC(gamma = 0.001, C= 100)

def accuracy(m):
    train_data = digits.data[:m]
    train_target = digits.target[:m]
    test_data = digits.data[m:]
    test_target = digits.target[m:]

    n = len(test_target)
    clf.fit(train_data,train_target)
    return sum(clf.predict(test_data)==test_target)*1.0/n

l = len(digits.target)
m = range(100,l,10)
p = [accuracy(i) for i in m]
plt.plot(m,p)
```

```python
import numpy
def accuracy(m):
    train_data = digits.data[:m]
    permutated_train_data = numpy.random.permutation(train_data)
    train_target = digits.target[:m]
    test_data = digits.data[m:]
    test_target = digits.target[m:]

    n = len(test_target)
    clf.fit(permutated_train_data,train_target)
    return sum(clf.predict(test_data)==test_target)*1.0/n

l = len(digits.target)
m = range(100,l,10)
p = [accuracy(i) for i in m]
plt.plot(m,p)
```
