```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import scipy.integrate as spint
```

```python
def f(z,t, mu):
    dz = np.zeros(len(z))
    dz[0] = z[1]
    dz[1] = mu*(1-z[0]**2)*z[1] - z[0]
    return dz
```

```python
mu = 2
f1 = lambda z,t: f(z,t,mu)
f1([1,1],2)
t = np.linspace(0, 10, 100)
z = spint.odeint(f1, np.array([1,1]), t)
```

```python
plt.plot(z[:,0])
```

```python
plt.plot(z[:,1])
```

```python
plt.plot(z[:,0], z[:,1])
```
